```python
#Bài 6
def prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
def sum_prime(a, b):
    total = 0
    for i in range(a, b + 1):
        if prime(i):
            total += i
    return total
while True:
    try:
        a, b = map(int, input().split())
        print(sum_prime(a, b))
    except:
        break
```

```python
#Bài 7*
n = int(input("Nhập n (n>=2): "))
i = 2
lonnhat = 1
while i * i <= n:
    while n % i == 0:
        lonnhat = i
        n //= i
    i += 1
if n > 1:
    lonnhat = n
print("Ước số nguyên tố lớn nhất:", lonnhat)
```

```python
#Bài 8
def reverse(n):
    return int(str(n)[::-1])
def Palindrome(n):
    return str(n) == str(reverse(n))
n = int(input("nhập n: "))
step = 0
while not Palindrome(n):
    n = n + reverse(n)
    step += 1
print(step)
print(n)
```

```python
#Bài 20
n = int(input("nhập n: "))
def chia(n):
    while n > 1:
        if n % 2 == 0:
            n //= 2
        else:
            break
    return n
if chia(n) == 1:
    print(f"{n} có dạng 2^k ")
else:
    print(f"{n} không có dạng 2^k ")
```
