```python
#  W7A1
ds = list(map(int, input().split()))
n = int(input())
def binary_search(ds):
    if n in ds:
        return ds.index(n)
    else:
        return -1
print(binary_search(ds))
```

```python
#  W7A2
def count_occurrences(arr, x) -> int:
    step = 0
    for i in arr:
        if i == x:
            step += 1
    return step
arr = list(map(int, input().split()))
x = int(input())
print(count_occurrences(arr, x))
```
